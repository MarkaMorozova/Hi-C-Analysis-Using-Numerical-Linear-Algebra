```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_array(filename):
    array = np.load(filename)
    return array



def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0: 
        return v
    return v / norm



bG = pd.read_csv('RNA-seq.BinSize.50kb.bedgraph', sep = '\t', skiprows=1, header=None)
bG[3] = bG[3].replace('NAN',0)
bG = bG[bG[0] == 'chr2']
bG[3] = bG[3].apply(float)
explog = np.log(bG[3] + 1)

plt.hist(explog,bins= 50,)
plt.title('Distribution of bins activity units (log)')
plt.grid()


GNM_array = read_array('mobility_profile.npy')
GNM_stat = read_array('status_arr.npy')
GNM_stat = (GNM_stat + 1)/2
GNM_array_norm = normalize(GNM_array-np.mean(GNM_array))
explog_norm = normalize(explog-np.mean(explog))


plt.scatter(GNM_array_norm,explog_norm)
plt.xlabel('GNM evaluation')
plt.ylabel('Bins activity')


fig, axs = plt.subplots(2, 1)
axs[0].plot(list(range(500,1000)),np.convolve(GNM_array_norm, np.ones(25)/25, mode='valid')[500:1000])
axs[0].set_xlabel('Time')
axs[0].set_ylabel('GNM evaluation')
axs[0].grid(True)
axs[1].plot(list(range(500,1000)),np.convolve(explog, np.ones(25)/25, mode='valid')[500:1000])
axs[1].set_ylabel('Experimental Activity')
axs[1].grid(True)

fig.tight_layout()
plt.show()


eigen_array = pd.read_csv('1_eigenvector_values.csv', sep = '\t', header=None)[0]


plt.scatter(eigen_array,explog,alpha=0.5)
plt.xlabel('Eigenvector')
plt.ylabel('Bins activity')plt.scatter(eigen_array,explog,alpha=0.5)
plt.xlabel('Eigenvector')
plt.ylabel('Bins activity')

fig, axs = plt.subplots(2, 1)
axs[0].plot(list(range(500,1000)),eigen_array[500:1000])
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Eigenvector')
axs[0].grid(True)
axs[1].plot(list(range(500,1000)),np.convolve(explog, np.ones(25)/25, mode='valid')[500:1000])
axs[1].set_ylabel('Experimental Activity')
axs[1].grid(True)

fig.tight_layout()
plt.show()

x1 = explog[GNM_stat == 1 ]
x2 = explog[GNM_stat == 0]
plt.violinplot([x2,x1])
plt.title('Difference in activity depending on GNM')
#xtickNames = plt.setp(xticklabels=)
plt.xlabel('Eigenvector < 0                                       Eigenvector > 0')
```
